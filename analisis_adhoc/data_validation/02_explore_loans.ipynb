Objetivo: Explorar y entender el dataset de pr√©stamos
- ¬øCu√°ntos pr√©stamos hay?
- ¬øPor qu√© hay m√∫ltiples filas por pr√©stamo? (snapshots)

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
base_path = Path.cwd()
while 'analisis_adhoc' in str(base_path):
    base_path = base_path.parent

data_path = base_path / 'data' / 'exports'

print(f"üìÅ Working dir: {Path.cwd()}")
print(f"üìÇ Base path: {base_path}")
print(f"üìç Data path: {data_path}")
print(f"‚úÖ Existe? {data_path.exists()}")

loans = pd.read_excel(data_path / 'loans.xlsx')
print(f"‚úÖ Loans: {len(loans):,}")
```

```python

print("="*80)
print("üìä DATASET: LOANS")
print("="*80)

# 1. ¬øCu√°ntas filas hay?
print(f"\n1Ô∏è‚É£ Total de filas en el CSV: {len(loans):,}")

# 2. ¬øQu√© columnas tengo?
print(f"\n2Ô∏è‚É£ Columnas disponibles:")
for col in loans.columns:
    print(f"   - {col}")

# 3. Ver primeros registros
print(f"\n3Ô∏è‚É£ Primeras 10 filas:")
display(loans.head(10))

# 4. ¬øHay datos faltantes?
print(f"\n4Ô∏è‚É£ Valores nulos por columna:")
print(loans.isnull().sum())
```

```python

# 5. üîç PROBLEMA: ¬øCu√°ntos pr√©stamos √öNICOS hay?
print(f"\n5Ô∏è‚É£ AN√ÅLISIS DE DUPLICADOS:")
unique_loans = loans['loan_id'].nunique()
total_rows = len(loans)
print(f"   - Pr√©stamos √∫nicos (loan_id): {unique_loans:,}")
print(f"   - Total de filas: {total_rows:,}")
print(f"   - Diferencia: {total_rows - unique_loans:,} filas adicionales")
print(f"   - Promedio de filas por pr√©stamo: {total_rows / unique_loans:.1f}")
```

```python
# 6. An√°lisis de un pr√©stamo espec√≠fico
ejemplo_loan = loans['loan_id'].iloc[0]
ejemplo_df = loans[loans['loan_id'] == ejemplo_loan]

print(f"\n6Ô∏è‚É£ EJEMPLO: loan_id = {ejemplo_loan}")
print(f"   Este pr√©stamo aparece {len(ejemplo_df)} vez(ces) en el CSV")

if len(ejemplo_df) == 1:
    print("\n   Informaci√≥n del pr√©stamo:") 
    ejemplo = ejemplo_df.iloc[0]
    cols_mostrar = ['user_id', 'vintage_date', 'Total Loan Amount', 'interest_rate', 
                    'loan_term', 'risk_band_production_customer', 'delinquency_status', 
                    'net_profit', 'capital_balance']
    
    for col in cols_mostrar:
        if col in loans.columns:
            print(f"   - {col}: {ejemplo[col]}")

print(f"\nüí° INTERPRETACI√ìN:")
print(f"   - Cada FILA = 1 PR√âSTAMO individual completo")
print(f"   - vintage_date = fecha en que se origin√≥ el pr√©stamo")
print(f"   - Los valores financieros son ACUMULADOS para el ciclo de vida del pr√©stamo")

```

```python
# 7. Vintage Month DIstribution (mes de originaci√≥n)
print(f"\n7Ô∏è‚É£ PR√âSTAMOS por mes de originaci√≥n (vintage_month):\"")
vintage_counts = loans['vintage_month'].value_counts().sort_index()
print(vintage_counts)
```

```python
# 8. ¬øCu√°ndo se desembolsaron los pr√©stamos? (vintage)
loans['vintage_date'] = pd.to_datetime(loans['vintage_date'])
loans['vintage_month'] = loans['vintage_date'].dt.to_period('M')

print(f"\n8Ô∏è‚É£ VINTAGES (mes de originaci√≥n):\"")
print(loans['vintage_month'].value_counts().sort_index())
```
